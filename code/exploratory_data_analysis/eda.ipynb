```python
import pandas as pd

pd.set_option('display.max_columns', 100)

df_train = pd.read_parquet('../../data/model_data/training_data.parquet').sort_values(by=['Date', 'Ticker'], ignore_index=True)

df_train
```

```python
df_train['Target'] = df_train['Adj Close in 60 Days'] - df_train['Adj Close']

df_train
```

```python
df_train.describe()
```

```python
df_train.corr(numeric_only=True)[['Adj Close in 60 Days', 'Target']].sort_values(by=['Target'], ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.boxplot(
    data=df_train.assign(f=pd.cut(df_train['Congressional Sales Past 90 Calendar Days'], bins=[0, 1, 5, 10, 15, 20, 50])),
    x='f',
    y='Target'
)

plt.ylim(-100, 100)

plt.show()
```
