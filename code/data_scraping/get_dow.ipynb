```python
import yfinance as yf

dow_jones_tickers = [
    "AAPL", "AMGN", "AXP", "BA", "CAT", "CRM", "CSCO", "CVX", "DIS", "DOW",
    "GS", "HD", "HON", "IBM", "INTC", "JNJ", "JPM", "KO", "MCD", "MMM",
    "MRK", "MSFT", "NKE", "PG", "TRV", "UNH", "V", "VZ", "WBA", "WMT"
]

df = yf.download(dow_jones_tickers, start='2015-10-01', end='2024-09-30').stack(level='Ticker').reset_index()

df.columns.name = None

df
```

```python
df.info()
```

```python
df.to_parquet('../../data/datasets/daily_stock_prices.parquet')

print('Done')
```
