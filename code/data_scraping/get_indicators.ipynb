### Unemployment Rates

```python
import pandas_datareader as pdr

df_unemployment = pdr.get_data_fred('UNRATE', start='2015-07-01', end='2024-09-30').reset_index()

df_unemployment.columns = ['Date', 'Unemployment Rate']

df_unemployment
```

```python
df_unemployment.info()
```

```python
df_unemployment.to_parquet('../../data/datasets/unemployment.parquet')

print('Done')
```
